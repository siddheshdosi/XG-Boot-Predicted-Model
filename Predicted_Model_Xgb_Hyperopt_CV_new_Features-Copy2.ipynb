```python
def data_fill(df):
    #Remove row which have 'for' in Average_Cost
    df = df[df.Average_Cost!='for']
    #Conver into int
    
    df['Average_Cost'] = df['Average_Cost'].apply(lambda x : int(x.replace(',','').replace('₹','')))
    df['Minimum_Order'] = df['Minimum_Order'].apply(lambda x : int(x.replace(',','').replace('₹','')))
    
    df['log_avg_cost'] = np.log(df['Average_Cost'])
    df['log_min_order'] = np.log(df['Minimum_Order'])
    
#     df['Rating_Bins'] = df['Rating'].apply(lambda x : x if x in ['Opening Soon','NEW','-','Temporarily Closed']
#                                                       else 'Rating_<=1.5' if float(x)<=1.5
#                                                       else 'Rating_1.5-2.0' if float(x)>1.5 and float(x)<=2.0 
#                                                       else 'Rating_2.0-2.5' if float(x)>2.0 and float(x)<=2.5
#                                                       else 'Rating_2.5-3.0' if float(x)>2.5 and float(x)<=3.0
#                                                       else 'Rating_3.0-3.5' if float(x)>3.0 and float(x)<=3.5
#                                                       else 'Rating_3.5-4.0' if float(x)>3.5 and float(x)<=4.0
#                                                       else 'Rating_>4.0' if float(x)>4.0 else x)
#     df['new_rating'] = df['Rating'].apply(lambda x : -1 if x in ['Opening Soon','NEW','-','Temporarily Closed']
#                                                       else float(x))
    
    
    df.loc[df.Votes=='-','Votes'] = -1
    df.loc[df.Reviews=='-','Reviews']=-1
    
    df['Votes'] = df.Votes.astype('int')
    df['Reviews'] = df.Reviews.astype('int')
    
    df['Votes-Reviews'] = np.log(df['Votes']-df['Reviews'])
    
    df['TOT_Rating_By_Votes']=df['Rating'].apply(lambda x : -1 if x in ['Opening Soon','NEW','-','Temporarily Closed'] else float(x))
    df['TOT_Rating_By_Votes'] = np.log(np.where(df['TOT_Rating_By_Votes']==-1,-1,df['TOT_Rating_By_Votes']*df['Votes']))
    
    df['TOT_Rating_By_Review']=df['Rating'].apply(lambda x : -1 if x in ['Opening Soon','NEW','-','Temporarily Closed'] else float(x))
    df['TOT_Rating_By_Review'] = np.log(np.where(df['TOT_Rating_By_Review']==-1,-1,df['TOT_Rating_By_Review']*df['Reviews']))
    
    df['TOT_Rating_By_Vote_Review']=df['Rating'].apply(lambda x : -1 if x in ['Opening Soon','NEW','-','Temporarily Closed'] else float(x))
    df['TOT_Rating_By_Vote_Review'] = np.log(np.where(df['TOT_Rating_By_Vote_Review']==-1,-1,df['TOT_Rating_By_Vote_Review']*(df['Reviews']+df['Votes'])))
    
    
    df['TOT_Rating_By_Vote-Review']=df['Rating'].apply(lambda x : -1 if x in ['Opening Soon','NEW','-','Temporarily Closed'] else float(x))
    df['TOT_Rating_By_Vote-Review'] = np.log(np.where(df['TOT_Rating_By_Vote-Review']==-1,-1,df['TOT_Rating_By_Vote_Review']*(df['Votes'])-df['Reviews']))


    df['NUMBER_OF_Cuisines'] = df['Cuisines'].apply(lambda x : len(x.split(',')))
    
    df['AVG_TOT_COST'] = df['Average_Cost']*df['Votes']
    df['AVG_TOT_COST'] = df['AVG_TOT_COST'].apply(lambda x : -1 if x<0 else x)
    
    df['AVG_TOT_COST_Reviews'] = df['Average_Cost']*df['Reviews']
    df['AVG_TOT_COST_Reviews'] = df['AVG_TOT_COST_Reviews'].apply(lambda x : -1 if x<0 else x)
    
    df['AVG_TOT_COST_Votes-Reviews'] = df['Average_Cost']*(df['Votes']-df['Reviews'])
    df['AVG_TOT_COST_Votes-Reviews'] = df['AVG_TOT_COST_Votes-Reviews'].apply(lambda x : -1 if x<0 else x)
    
    df['AVG_TOT_MIN_ORDER'] = df['Minimum_Order']*df['Votes']
    df['AVG_TOT_MIN_ORDER'] = df['AVG_TOT_MIN_ORDER'].apply(lambda x : -1 if x<0 else x)
    
    df['AVG_TOT_MIN_ORDER'] = df['Minimum_Order']*df['Votes']
    df['AVG_TOT_MIN_ORDER'] = df['AVG_TOT_MIN_ORDER'].apply(lambda x : -1 if x<0 else x)
    
    df['AVG_TOT_MIN_ORDER_Review'] = df['Minimum_Order']*df['Reviews']
    df['AVG_TOT_MIN_ORDER_Review'] = df['AVG_TOT_MIN_ORDER_Review'].apply(lambda x : -1 if x<0 and x==0 else x)
    
    df['AVG_TOT_MIN_ORDER_Votes-Review'] = df['Minimum_Order']*(df['Votes']-df['Reviews'])
    df['AVG_TOT_MIN_ORDER_Votes-Review'] = df['AVG_TOT_MIN_ORDER_Review'].apply(lambda x : -1 if x<0 and x==0 else x)
    
    #df = pd.get_dummies(df,columns=['Rating_Bins'])
    df = pd.get_dummies(df,columns=['Location'])
    return df
```

```python
###Hyper opt XGB
import xgboost as xgb
import pandas as pd
import numpy as np
from uuid import uuid4
from sys import exit
from hyperopt.pyll.base import scope
from hyperopt.pyll.stochastic import sample
from hyperopt import STATUS_OK, Trials, fmin, hp, tpe
import random
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, roc_curve, auc,log_loss
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedShuffleSplit


#from sklearn import metrics

# stratSplit = StratifiedShuffleSplit(new_y, test_size=0.25, random_state=42)
# for train_idx, test_idx in stratSplit.split:
#     X_train=X[train_idx]
#     y_train=y[train_idx]
#     X_val=X[test_idx]
#     y_val = y[test_idx]
X_train, X_val, y_train, y_val = train_test_split(X, new_y, random_state=42, test_size=0.25,stratify=new_y)

#print(y_train.value_counts())
#print(y_val.value_counts())

num_train, num_feature = X_train.shape

xgb_train = xgb.DMatrix(X_train.values, y_train.values)
xgb_eval  = xgb.DMatrix(X_val.values, y_val.values)


x_features = X.columns
test= test_data_pre[x_features]

xgb_test   = xgb.DMatrix(test.values)


###Function to get imp features from xgb
def ceate_feature_map(features):
    outfile = open('xgb.fmap', 'w')
    i = 0
    for feat in features:
        outfile.write('{0}\ti\t{1}\tq\n'.format(i, feat))
        i = i + 1

    outfile.close()


import itertools
N_HYPEROPT_PROBES = 50000 #change to 5000
EARLY_STOPPING = 80 #change to 80
HOLDOUT_SEED = 123456
HOLDOUT_SIZE = 0.10
HYPEROPT_ALGO = tpe.suggest  #  tpe.suggest OR hyperopt.rand.suggest
SEED0 = random.randint(1,1000000000)
NB_CV_FOLDS = 5 #chagne to 5

obj_call_count = 0
cur_best_score = 0.55

import sklearn

def objective(space):
    
    global obj_call_count, cur_best_score, X_train, y_train, test, X_val, y_val

    
    obj_call_count += 1
    print('\nXGBoost objective call #{} cur_best_score={:7.5f}'.format(obj_call_count,cur_best_score) )

    sorted_params = sorted(space.items(), key=lambda z: z[0])
    print('Params:', str.join(' ', ['{}={}'.format(k, v) for k, v in sorted_params if not k.startswith('column:')]))


    xgb_params = sample(space)
       
    
    mdl = xgb.cv(
                        xgb_params,
                        xgb_train,
                        num_boost_round = 750,##change to 750,
                        nfold=NB_CV_FOLDS,
                        #metrics='binary_logloss',
                        stratified=True,
                        #fobj=None,
                        #feval=None,
                        #init_model=None,
                        #feature_name='auto',
                        early_stopping_rounds=EARLY_STOPPING,
                        #fpreproc=None,
                        verbose_eval=100,
                        show_stdv=False,
                        )

           
    #print('mdl : ',mdl)
    n_rounds = len(mdl["test-mlogloss-mean"])
    cv_score = mdl["test-mlogloss-mean"][n_rounds-1]

    print( 'CV finished n_rounds={} cv_score={:7.5f}'.format( n_rounds, cv_score ) )
    
    gbm_model = xgb.train(
                        xgb_params,
                        xgb_train,
                        num_boost_round=n_rounds,
                        #metrics='mlogloss',
                        # valid_names=None,
                        # fobj=None,
                        # init_model=None,
                        # feature_name='auto',
                        # categorical_feature='auto',
                        # early_stopping_rounds=None,
                        # evals_result=None,
                        verbose_eval=False
                        # learning_rates=None,
                        # keep_training_booster=False,
                        # callbacks=None)
                    )
    
    predictions = gbm_model.predict(xgb_eval,
                                    ntree_limit =n_rounds)
   
    #print(gbm_model.feature_importances)
#     ceate_feature_map(test_data_pre[x_features].columns)
#     importance = gbm_model.get_fscore(fmap='xgb.fmap')
#     importance = sorted(importance.items(), key=operator.itemgetter(1))

#     f_imp = pd.DataFrame(importance, columns=['feature', 'fscore'])
#     f_imp['fscore'] = f_imp['fscore'] / f_imp['fscore'].sum()
#     print(f_imp.head())

    
    #score = roc_auc_score(y_val, predictions)
    score = log_loss(y_val, predictions)
    print('valid score={}'.format(score))
    
    
    do_submit = False

    if score < cur_best_score:
        cur_best_score = score
        print('NEW BEST SCORE={}'.format(cur_best_score))
        do_submit = True

    if do_submit:
        do_submit=False
        submit_guid = uuid4()

        print('Compute submissions guid={}'.format(submit_guid))

        y_submission = gbm_model.predict(xgb_test, ntree_limit = n_rounds)
        
        import pickle
        f = open('xg_submission_5/new_Model_xgboost_score={:13.11f}_submission_guid={}.pkl'.format(score,submit_guid),'wb')
        pickle.dump(gbm_model, f)
        f.close()
        
        submission_filename = 'xg_submission_5/new_submission_file_xgboost_score={:13.11f}_submission_guid={}.xlsx'.format(score,submit_guid)
        df = pd.DataFrame()
        y_submission_2 = []
#         print(type(y_submission))
#         print(y_submission[0])
        for pred in y_submission:
            y_submission_2.append(np.argmax(pred))
            #print(np.argmax(pred))
        df['Delivery_Time'] = y_submission_2
        df['Delivery_Time'] = df['Delivery_Time'].apply(lambda x :label_dic_rev[x])
        df.to_excel(submission_filename, index=False)
        
        ### Calculate the feature imp
        #feature_importance_df = get_xgb_imp(gbm_model, test_data_pre[x_features].columns)
        #feature_importance_df.to_csv('xg_submission_5/Feature_Imp_xgboost_score={:13.11f}_submission_guid={}.csv'.format(score,submit_guid))
       
    loss = score
    #loss = metrics.log_loss(y, dtrainPredProb)
    return {'loss': loss, 'status': STATUS_OK}


# 'learning_rate':    hp.choice('learning_rate',    np.arange(0.05, 0.31, 0.05)),
# 'max_depth':        hp.choice('max_depth',        np.arange(5, 16, 1, dtype=int)),
# 'min_child_weight': hp.choice('min_child_weight', np.arange(1, 8, 1, dtype=int)),
# 'colsample_bytree': hp.choice('colsample_bytree', np.arange(0.3, 0.8, 0.1)),
# 'subsample':        hp.uniform('subsample', 0.8, 1),
# 'n_estimators':     100,

space ={
    'booster '    : 'gbtree',  
    #'num_class' : '6',
    'objective'   : 'multi:softprob',
    'eval_metric' : 'mlogloss',
    #'num_class' : (0,6],
   'max_depth':        hp.choice('max_depth',        np.arange(7, 25, 1, dtype=int)),  
   'colsample_bytree': hp.choice('colsample_bytree', np.arange(0.3, 0.8, 0.1)),
    'alpha'       : hp.uniform('alpha', 1e-4, 1e-6 ),
    'lambda'      : hp.uniform('lambda', 1e-4, 1e-6 ),
    
    'min_child_weight ': hp.loguniform('min_sum_hessian_in_leaf', 0, 2.3),
    'learning_rate':    hp.choice('learning_rate',    np.arange(0.05, 0.31, 0.05)),
    'subsample':        hp.uniform('subsample', 0.8, 1),
    'n_estimators':     100,
    'gamma': hp.uniform('gamma', 0.0, 0.5),
    'seed'             : hp.randint('seed',2000000),
    'num_class' : 6
   }


trials = Trials()
best = fmin(fn=objective,
                     space=space,
                     algo=HYPEROPT_ALGO,
                     max_evals=N_HYPEROPT_PROBES,
                     trials=trials
                   # ,verbose=1
                   )

print('-'*50)
print('The best params:')
print( best )
print('\n\n')
```
